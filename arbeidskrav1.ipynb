```python
import pandas as pd
import sympy as sp
from matplotlib import pyplot as plt
import numpy as np
df=pd.read_csv("https://uit-sok-1006-v23.github.io/data/suits_watermelons.csv",delimiter=";")

df["p"] = np.exp(df["log p"])
df["q"] = np.exp(df["log q"])
df["h"] = np.exp(df["log h"])

plt.plot(df["Year"], df["p"], label ="Price")
plt.xlabel("Year")
plt.ylabel("Price")
plt.legend()
```

```python
# 7.2, A, B


p, q = sp.symbols('p, q', real=True, positive=True)
eq_s = sp.Eq(-2 + 1.7*q, p)
eq_d = sp.Eq(8.5 - 0.8*q, p)
sol_s = sp.solve(eq_s, q)
sol_d = sp.solve(eq_d, q)
display(sol_s, sol_d)

#Vi beregner elastisitet for supply ved å løse ligningen med hensyn på Log q, Vi gjør det samme med demand

#Elastisiteten for supply er 0.588 som betyr at 1 % økning i etterspørsel fører til 0.588% økning i tilbud

#Elastisiteten for demand er -1.25 som betyr at 1 % økning i pris fører til en 1.25% nedgang i etterspørsel

#koeffisient variablen i tabellen viser et estimat på hvor mye etterspørselen endres dersom tilbudet øker med 1%

#I den andre tabllen viser hvor mye etterspørslen påvirkes dersom prisen øker med 1%


#Et shock er en ekstern påvirkning på markedet, et eksempel på et shock, kan være naturkatostrofer, pandemi el
```
